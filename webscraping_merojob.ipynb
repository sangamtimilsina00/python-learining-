```python
import numpy as np
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome()
for j in range(1,9):
    driver.get(f"https://merojob.com/category/accounting-finance?limit=6&offset={j}")
    time.sleep(10)
    for i in range(2,8):
        try:
            title=driver.find_element(By.XPATH,f"/html/body/div/div/main/div/div[2]/div/div[2]/div/div[2]/div[3]/div[1]/div[{i}]/div/div[1]/div/div[2]")
            print(title.text)
        except:
            print("data not found")
        try:
            job_des=driver.find_element(By.XPATH,f"/html/body/div/div/main/div/div[2]/div/div[2]/div/div[2]/div[3]/div[1]/div[{i}]/div/div[2]/div[1]")
            print(job_des.text)
        except:
            print("data not found")
        try:
            job_salary=driver.find_element(By.XPATH,f"/html/body/div/div/main/div/div[2]/div/div[2]/div/div[2]/div[3]/div[1]/div[{i}]/div/div[2]/div[2]")
            print(job_salary.text)
        except:
            print("data not found")
        try:
            job_skills=driver.find_element(By.XPATH,f'/html/body/div/div/main/div/div[2]/div/div[2]/div/div[2]/div[3]/div[1]/div[{i}]/div/div[2]/div[3]')
            print(job_skills.text)
        except:
            print("data not found")
        print("#"*100)
```
