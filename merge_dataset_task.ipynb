```python
jan_temp=pd.merge(courses,jan,on="course_id",how="outer")
feb_temp=pd.merge(courses,feb,on="course_id",how="inner")

total_feb=feb_temp["price"].sum()
total_jan=jan_temp['price'].sum()
total=total_feb+total_jan
print("Total revenue :Rs",total)
print("Total revenu in jan is Rs",total_jan)
print("Total revenue in feb is Rs ",total_feb)
```
