```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
driver=webdriver.Chrome()
```

```python
# /html/body/div[1]/div[2]/div/section[5]/div/div/div/div/div/div/div[1]/div[1]/div/div[3]/h3/a
# /html/body/div[1]/div[2]/div/section[5]/div/div/div/div/div/div/div[1]/div[2]/div/div[3]/h3/a
# /html/body/div[1]/div[2]/div/section[5]/div/div/div/div/div/div/div[1]/div[3]/div/div[3]/h3/a

# old pricce
# /html/body/div[1]/div[2]/div/section[5]/div/div/div/div/div/div/div[1]/div[1]/div/div[4]/div[1]/div[1]/span/del/span/bdi
# /html/body/div[1]/div[2]/div/section[5]/div/div/div/div/div/div/div[1]/div[3]/div/div[4]/div/div[1]/span/del/span/bdi
# /html/body/div[1]/div[2]/div/section[5]/div/div/div/div/div/div/div[1]/div[4]/div/div[4]/div/div[1]/span/del/span/bdi

# newprice
# /html/body/div[1]/div[2]/div/section[5]/div/div/div/div/div/div/div[1]/div[1]/div/div[4]/div[1]/div[1]/span/ins/span/bdi
# /html/body/div[1]/div[2]/div/section[5]/div/div/div/div/div/div/div[1]/div[2]/div/div[4]/div/div[1]/span/ins/span/bdi
# /html/body/div[1]/div[2]/div/section[5]/div/div/div/div/div/div/div[1]/div[13]/div/div[4]/div/div[1]/span/ins/span/bdi
```

```python
loc="/html/body/div[1]/div[2]/div/section[5]/div/div/div/div/div/div/div[1]/div[1]/div/div[3]/h3/a"
obj=driver.find_element(By.XPATH,loc)
obj.text
```

```python
course_title=[]
old_price=[]
new_price=[]
for i in range(1,14):
    loc=f"/html/body/div[1]/div[2]/div/section[5]/div/div/div/div/div/div/div[1]/div[{i}]/div/div[3]/h3/a"
    obj=driver.find_element(By.XPATH,loc)
    course_title.append(obj.text)

    loc=f"/html/body/div[1]/div[2]/div/section[5]/div/div/div/div/div/div/div[1]/div[{i}]/div/div[4]/div[1]/div[1]/span/del/span/bdi"
    obj=driver.find_element(By.XPATH,loc)
    old_price.append(obj.text)

    loc=f"/html/body/div[1]/div[2]/div/section[5]/div/div/div/div/div/div/div[1]/div[{i}]/div/div[4]/div[1]/div[1]/span/ins/span/bdi"
    obj=driver.find_element(By.XPATH,loc)
    new_price.append(obj.text)
```

```python
dic={"Course":course_title,
     "Old Price":old_price,
     "New Price":new_price}
df=pd.DataFrame(dic)
```
